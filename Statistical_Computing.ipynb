```r
set.seed(3685)
x<-rnorm(20, mean=52, sd=4)
x_bar<-mean(x)
mu0<-50
sd<-sd(x)                 #سییگما نامعلوم
n<-length(x)
alpha<-0.05
# آماره آزمون
test_stat<- (x_bar - mu0)/(sd/sqrt(n))
# مقدار بحرانی برای زمانی که دو طرفه باشد.
df<-n-1
p_value<-2*(1-pt(abs(test_stat), df))
test_stat
p_value
###################### استفاده از کتابخانه زمانی که سیگما نامعلوم  باشد.
set.seed(3685)
x <- rnorm(20, mean = 52, sd = 4)

t.test(x, mu = 50, alternative = "two.sided")
```
