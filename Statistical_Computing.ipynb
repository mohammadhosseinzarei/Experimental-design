σ نامعلوم

```r
set.seed(3685)
x<-rnorm(20, mean=52, sd=4)
x_bar<-mean(x)
mu0<-50
sd<-sd(x)                 #سییگما نامعلوم
n<-length(x)
alpha<-0.05
# آماره آزمون
test_stat<- (x_bar - mu0)/(sd/sqrt(n))
# مقدار بحرانی برای زمانی که دو طرفه باشد.
df<-n-1
p_value<-2*(1-pt(abs(test_stat), df))
test_stat
p_value
# CI
t_critl<-qt(1-alpha/2, df)
CI_lower<-xbar-(t_critl*(sd/sqrt(n)))
CI_upper<-xbar+(t_critl*(sd/sqrt(n)))
CI_lower
CI_upper
###################### استفاده از کتابخانه زمانی که سیگما نامعلوم  باشد.
set.seed(3685)
x <- rnorm(20, mean = 52, sd = 4)

t.test(x, mu = 50, alternative  = "two.sided")


```

```r
set.seed(258)
x<-rnorm(20, mean=52, sd=4)
mu0<-50
xbar<-mean(x)
sigma<-4
n<-length(x)
alpha<-0.05
# آماره آزمون
z_stat=(xbar-mu0)/(sigma/sqrt(n))
#p_value
p_value<-2*(1-pnorm(z_stat))
z_stat
p_value
#CI
z_crit<-qnorm(1-alpha/2)
CI_lower=xbar-(z_crit*sigma/sqrt(n))
CI_upper=xbar+(z_crit*sigma/sqrt(n))
CI_lower
CI_upper
#use library
library(BSDA)
z.test(x, mu=50, sigma.x=4, alternative=("two.sided"))
```
